```python
import numpy as np
import time
import matplotlib.pyplot as plt

# Функция для перемножения двух матриц и измерения времени выполнения
def multiply_matrices_and_measure_time(size):
    matrix1 = np.random.rand(size, size)
    matrix2 = np.random.rand(size, size)
    start_time = time.time()
    result = np.dot(matrix1, matrix2)
    end_time = time.time()

    return end_time - start_time

# Размеры матриц
matrix_sizes = [10, 50, 100, 200, 300]
num_measurements = 10  # Количество измерений для усреднения

# Список для хранения результатов времени выполнения
execution_times = []

# Цикл для измерения времени выполнения операций и усреднения результатов
for size in matrix_sizes:
    measurements = []
    for _ in range(num_measurements):
        execution_time = multiply_matrices_and_measure_time(size)
        measurements.append(execution_time)

    avg_execution_time = np.mean(measurements)
    execution_times.append(avg_execution_time)

# Вывод графика зависимости среднего времени от размеров матрицы
plt.plot(matrix_sizes, execution_times, marker='o')
plt.xlabel('Размер матрицы')
plt.ylabel('Среднее время выполнения (сек)')
plt.title('График зависимости среднего времени выполнения от размера матрицы')
plt.grid()
plt.show()

```
